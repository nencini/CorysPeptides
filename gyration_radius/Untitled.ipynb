```python
data={}
for j,file in enumerate(os.listdir(folder_path)):
    data[j]={}
    x=[]
    y=[]
    data[j]["name"]=file
    with open("gyrate_nojump_"+file+".xvg") as f:
        for i,line in enumerate(f):
            if "#" in line:
                pass
            elif "@" in line:
                pass
            elif i%1==0:
                x.append(float(line.split()[0]))
                y.append(float(line.split()[1]))
    
    data[j]["x"]=x
    data[j]["y"]=y
    
    plt.rcParams["figure.figsize"] = [15.00, 7]
    plt.rcParams["figure.autolayout"] = True
    plt.rcParams.update({'font.size': 20})
    plt.title(file)

    plt.plot(x,y)
    plt.show()
                
```

```python
plt.rcParams["figure.figsize"] = [15.00, 8]
plt.rcParams["figure.autolayout"] = True
plt.rcParams.update({'font.size': 20})
fig, axs = plt.subplots(3,2)
for key in data:
    ax_x=key//2
    ax_y=key%2
    axs[ax_x][ax_y].plot(data[key]["x"],data[key]["y"])
    axs[ax_x][ax_y].title.set_text(data[key]["name"][14:20])
    axs[ax_x][ax_y].set_ylabel("r")
    axs[ax_x][ax_y].set_xlabel("time")
```
